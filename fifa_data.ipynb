Load the necessary libaries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Loading data

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa.head(5)
```

Histogram

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins, color='#3969b8')
plt.xticks(bins)
plt.ylabel('Numbers of players')
plt.xlabel('Skill Level')
plt.title('Distribution of Players skill in FIFA 2018')

plt.show()
```
