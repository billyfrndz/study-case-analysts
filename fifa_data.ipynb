Load the necessary libaries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Loading data

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa.head(5)
```

Histogram

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins, color='#3969b8')
plt.xticks(bins)
plt.ylabel('Numbers of players')
plt.xlabel('Skill Level')
plt.title('Distribution of Players skill in FIFA 2018')

plt.show()
```

Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].shape[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].shape[0]

labels = ['Left', 'Right']
colors = ['#5e8edb', '#4c72ed']
plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f%%')
plt.title('Foot Preferance of FIFA Players')


plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type (x) == str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].shape[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].shape[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].shape[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].shape[0]
heavy = fifa.loc[fifa.Weight > 200].shape[0]

weight = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', 'Under 125-150', 'Under 150-175', 'Under 175-200', 'Over 200']
explode = (.4,.2,0,0,.4)

plt.title('weight distribution of FIFA Players')
plt.pie(weight, labels=labels, autopct='%.2f%%', pctdistance=0.8, explode=explode)

plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
Madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels=['FC Barcelona', 'Real Madrid', 'NE Revolution']

boxes = plt.boxplot([barcelona, Madrid, revs], labels= labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    #set edge color
    box.set(color='#60f793', linewidth=2)

    #change fillcolor
    box.set(facecolor='#a1ffc0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('Fifa Overall Rating')
plt.show()
```
